```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
#reading from csv file 
ratings = pd.read_csv("./IndiaMovieDatabase/2010-2019/bollywood_ratings_2010-2019.csv")
ratings
```

```python
ratings.duplicated().sum()
```

```python
ratings.drop_duplicates('imdb_id' , inplace = True)
```

```python
ratings.isnull().sum()
```

```python
ratings = ratings.dropna()
```

```python
ratings.isnull().sum()
ratings
```

```python
#meta file
meta = pd.read_csv("./IndiaMovieDatabase/2010-2019/bollywood_meta_2010-2019.csv")
meta
```

```python
meta.duplicated('imdb_id').sum()
```

```python
meta.drop_duplicates('imdb_id' , inplace = True)
meta.duplicated().sum()
```

```python
meta.isnull().sum()
```

```python
meta
```

```python
#merging both the ratings and met file datasets
meta_ratings = pd.merge(meta, ratings,on = "imdb_id" )
meta_ratings
```

```python
print(meta_ratings.isnull().sum(), meta_ratings.duplicated().sum(), sep = "\n\n")
```

```python
#sorting to get the most rated movies at top
meta_ratings.sort_values(by="imdb_rating", ascending = False , inplace = True)
meta_ratings
```

```python
z =  meta_ratings[0:5]
z
```

```python
fig = plt.figure()
sns.barplot(x = 'imdb_rating' , y ='title'  , data= z)
```
