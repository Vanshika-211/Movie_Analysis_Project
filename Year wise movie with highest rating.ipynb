```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
#reading from csv file 
ratings = pd.read_csv("./IndiaMovieDatabase/2010-2019/bollywood_ratings_2010-2019.csv")
ratings
```

```python
ratings.duplicated().sum()
```

```python
ratings.drop_duplicates('imdb_id' , inplace = True)
```

```python
ratings.isnull().sum()
```

```python
ratings = ratings.dropna()
```

```python
ratings.isnull().sum()
ratings
```

```python
ratings.replace(0,np.nan ,  inplace = True)
ratings
```

```python
ratings.dropna(inplace = True)
ratings
```

```python
meta = pd.read_csv("./IndiaMovieDatabase/2010-2019/bollywood_meta_2010-2019.csv")
meta
```

```python
meta.isnull().sum()
```

```python
meta.drop_duplicates('imdb_id', inplace = True)
meta
```

```python
#merging both the ratings and met file datasets
meta_ratings = pd.merge(meta, ratings,on = "imdb_id" )
meta_ratings
```

```python
print(meta_ratings.isnull().sum(), meta_ratings.duplicated().sum(), sep = "\n\n")
```

```python
year = meta_ratings.groupby(meta_ratings['year_of_release'])
year
```

```python
for n , g in year :
    print (n,g , sep="\n" , end="\n\n")
```

```python
p = year.agg(max)
p
```

```python
fig = plt.figure()
sns.barplot(y =p.title  , x = p.imdb_rating )
```
